### Task A: Image Manipulation

1. Read the image Parrots.jpg
2. Show the two images of R and B channel only
3. Mirror the full image
4. Shrink the image into a new image n times smaller
5. Save into a new image the file of the shrunk picture

Task A

```python
# reading the image and receiving the shape (image matrix size/resolution)
P = plt.imread("Parrots.jpg")
Pshape = P.shape

print(type(Pshape)) # class
print(Pshape) # resolution

# display the image
plt.imshow(P)
plt.show()
```

The layer of the matrix indicates the colour, (R = 0, G = 1, B = 2)

```python
# show the RED channel only
# put the G and B channels to zero (1 and 2 = 0)
R = np.copy(P)
R[:,:,1] = 0
R[:,:,2] = 0

# display image
plt.imshow(R)
plt.show()
```

```python
# show the BLUE channel only
# put the R and G channels to zero (0 and 1 = 0)
B = np.copy(P)
B[:,:,0] = 0
B[:,:,1] = 0

# display image
plt.imshow(B)
plt.show()
```

```python
# show the GREEN channel only
# put the R and B channels to zero (0 and 2 = 0)
G = np.copy(P)
G[:,:,0] = 0
G[:,:,2] = 0

# display image
plt.imshow(G)
plt.show()
```

Mirroring images

```python
M = np.copy(P)

# traverse each column for each row and insert into the matrix starting from right to left
M = M[:, -1:-Pshape[1]-1:-1, :]

# display image
plt.imshow(M)
plt.show()
```

Shrinking images

```python
# shrink by N, i.e. N time smaller
N = 20
```

```python
# size will be 1/N of initial size (use integers for N)
# size of new shrunk array, remember to keep the layers (colours) the same
Sshape = (int(Pshape[0]/N),int(Pshape[1]/N),3)

# matrix solution, S of the desired shape
S = np.zeros(Sshape)
```

```python
# traverse each column for each row at every N values and insert into matrix
# for all layers
S = P[0:Pshape[0]:N, 0:Pshape[1]:N, :]

# display image, S
plt.imshow(S)

# save image
#plt.imsave('Shrunk.jpg', S)
```

### Task B: Vector Fields

1. Make an array of the desired range
2. Create a 2D grid to receive every combination of x and y
3. Make an ND array for the field, F
4. Compute function from the meshgrid and fill in array with F
5. Generate solution matrix
6. Compute final solution
7. Plot if needed

Sketching Vector Fields

```python
# define the range
x = np.arange(-5,5,0.5)
```

```python
Nx = len(x)
# create 2D grid
Xg , Yg = np.meshgrid(x,x)
```

```python
F = np.ndarray( (Nx,Nx,2) )
```

```python
# i component:
F[:,:,0] = Yg / (Xg**2+Yg**2)

# j component:
F[:,:,1] = -Xg / (Xg**2+Yg**2)
```

```python
# plot streamlines of vector field
plt.streamplot(Xg,Yg,F[:,:,0],F[:,:,1])
plt.show()

# plot the actual vectors
plt.quiver(Xg,Yg,F[:,:,0],F[:,:,1],color='b')
plt.show()
```
